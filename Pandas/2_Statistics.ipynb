```python
# Print the maximum of the date column
print(df['date'].max())
```

```python
# Print the info about the sales DataFrame
print(df.info())
# Print the mean of weekly_sales
print(df['weekly_sales'].mean())

# Print the median of weekly_sales
print(df['weekly_sales'].median())
```

```python
# Import NumPy and create custom IQR function
# IQR ->is short for inter-quartile range, which is the 75th percentile minus the 25th percentile
import numpy as np
def iqr(column):
    return column.quantile(0.75) - column.quantile(0.25)

# Update to print IQR and median of temperature_c, fuel_price_usd_per_l, & unemployment
print(df[["temperature_c", "fuel_price_usd_per_l", "unemployment"]].agg([iqr, np.median]))
```

```python
# Sort sales_1_1 by date
sales_1_1 = df.sort_values("date")
sales_1_1.head()
```

# Cumulative statistics

```python

# Get the cumulative sum of weekly_sales, add as cum_weekly_sales col
sales_1_1['cum_weekly_sales'] = sales_1_1['weekly_sales'].cumsum()
sales_1_1.head()
```

```python
# Get the cumulative max of weekly_sales, add as cum_max_sales col
sales_1_1['cum_max_sales']=sales_1_1['cum_weekly_sales'].cummax()
sales_1_1.head()
```

```python
# See the columns you calculated
print(sales_1_1[["date", "weekly_sales", "cum_weekly_sales", "cum_max_sales"]].head())
```
