# Pivot Table

```python
# Add a year column to temperatures
temperatures['year']=temperatures['date'].dt.year
temperatures.head()
```

```python
# Pivot avg_temp_c by country and city vs year
temp_by_country_city_vs_year = temperatures.pivot_table("avg_temp_c", columns="year", index=["country","city"], aggfunc='first')
temp_by_country_city_vs_year.head()
```

# Calculating on a pivot table

```python
# Get the worldwide mean temp by year
mean_temp_by_year = temp_by_country_city_vs_year.mean(axis="index")
mean_temp_by_year
```

```python
# Filter for the year that had the highest mean temp
print(mean_temp_by_year[mean_temp_by_year==mean_temp_by_year.max()])
```

```python
# Get the mean temp by city
mean_temp_by_city = temp_by_country_city_vs_year.mean(axis="columns")
mean_temp_by_city
```

```python
# Get the mean temp by city
mean_temp_by_city[mean_temp_by_city==mean_temp_by_city.min()]
```
